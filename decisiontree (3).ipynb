```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Step 2: Load the dataset
iris_data = load_iris()
features = iris_data.data
labels = iris_data.target

# Let's print some info
print("Shape of data:", features.shape)
print("Classes:", iris_data.target_names)

# Step 3: Train the Decision Tree Classifier on full data
tree_model = DecisionTreeClassifier(random_state=0)
tree_model.fit(features, labels)
print("The decision tree model has been trained.")
# Predicting using the trained model
predictions = tree_model.predict(features)
# Checking accuracy and classification report
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(labels, predictions)
print("Accuracy of the model:", round(accuracy * 100, 2), "%")

print("\nClassification Report:\n")
print(classification_report(labels, predictions, target_names=iris_data.target_names))

print("The decision tree model has been trained.")

# Step 4: Plot the trained decision tree
plt.figure(figsize=(14, 10))
plot_tree(
    tree_model,
    feature_names=iris_data.feature_names,
    class_names=iris_data.target_names,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree (Iris Dataset)")
plt.show()

```
